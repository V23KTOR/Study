# Binomial Distribution

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

n = 5   # Кол-во испытаний
p = 0.6 # Вероятность успеха
x = 3   # Кол-во успехов, которые мы ожидаем

print("Binomial Distribution Analysis")
print(f"P(X) = {p}")
print("-" * 55)


for x in range(n+1):

    # stats для биномиального PMF
    p_x  = round(stats.binom.pmf(x, n, p),2)
    # use the probability for future work
    p_xc = round(stats.binom.cdf(x, n, p),2)
    
    print(f'P(X = {x}) = {p_x}    P(X <={x}) = {p_xc}    P(X >{x}) = {round(1-p_xc,2)}')

print("-" * 55)

# Визуализация

fig,axes = plt.subplots(1,2,figsize=(10,3))
x_vals = np.arange(0, n+1)

pmf_vals = stats.binom.pmf(x_vals, n, p)
axes[0].bar(x_vals, pmf_vals, alpha=0.7)
axes[0].set_xlabel('Number of Successes')
axes[0].set_ylabel('Probability')
axes[0].set_title(f'Binomial Distribution (n={n}, p={p})',size=10)


pmf_valsc = stats.binom.cdf(x_vals, n, p)
axes[1].bar(x_vals, pmf_valsc, alpha=0.7, color='grey')
axes[1].set_xlabel('Number of Successes')
axes[1].set_ylabel('Probability')
axes[1].set_title(f'Binomial Cumulative  (<=) Distribution (n={n}, p={p})',size=10)

plt.show()
```

# Poisson Distribution

```python
# Parameters for Poisson distribution
lam = 3   # Средняя частота появления эффекта в заданный интервал!
max_x = 9  # Для какого кол-ва значений случайной величины хотип построить распределение

print("Poisson Distribution Analysis")
print(f"Lambda (λ) = {lam}")
print("-" * 55)

for x in range(max_x + 1):
    # stats для Poisson PMF и CDF
    p_x = round(stats.poisson.pmf(x, lam), 1)    # P(X = x)
    p_xc = round(stats.poisson.cdf(x, lam), 1)   # P(X <= x)
    
    print(f'P(X = {x}) = {p_x}    P(X <= {x}) = {p_xc}    P(X > {x}) = {round(1-p_xc, 1)}')


print("-" * 55)

# Визуализация

fig,axes = plt.subplots(1,2,figsize=(10,3))
x_vals = np.arange(0, max_x + 1)

pmf_vals = stats.poisson.pmf(x_vals, lam)
axes[0].bar(x_vals, pmf_vals, alpha=0.7)
axes[0].set_xlabel('Number of Event Occurence')
axes[0].set_ylabel('Probability')
axes[0].set_title(f'Poisson Distribution  (λ = {lam})',size=10)


pmf_valsс = stats.poisson.cdf(x_vals, lam)
axes[1].bar(x_vals, pmf_valsс, alpha=0.7, color='grey')
axes[1].set_xlabel('Number of Event Occurence')
axes[1].set_ylabel('Probability')
axes[1].set_title(f'Poisson Cumulative (<=) Distribution (λ = {lam})',size=10)

plt.show()
```

# Geometric Random Variable 
## (Number of experiments before success)

```python

# Parameters for Geometric distribution
p = 0.3   # 
max_x = 8  # Maximum number of trials to consider

print("Geometric Distribution Analysis")
print(f"Probability of success (p) = {p}")
print("-" * 65)

for x in range(1, max_x + 2):  
    
    p_x = round(stats.geom.pmf(x, p), 1)    # P(X = x)
    p_xc = round(stats.geom.cdf(x, p), 1)   # P(X <= x)
    
    print(f'P(X = {x}) = {p_x}    P(X <= {x}) = {p_xc}    P(X > {x}) = {round(1-p_xc, 1)}')

print("-" * 65)

# Визуализация
fig, axes = plt.subplots(1, 2, figsize=(10, 3))
x_vals = np.arange(1, max_x + 2)

pmf_vals = stats.geom.pmf(x_vals, p)
axes[0].bar(x_vals, pmf_vals, alpha=0.7, )
axes[0].set_xlabel('Number of Trials Until First Success')
axes[0].set_ylabel('Probability')
axes[0].set_title(f'Geometric Distribution PMF (p = {p})')
axes[0].grid(True, alpha=0.3)

# Cumulative distribution
cdf_vals = stats.geom.cdf(x_vals, p)
axes[1].bar(x_vals, cdf_vals, alpha=0.7, color='grey')
axes[1].set_xlabel('Number of Trials Until First Success')
axes[1].set_ylabel('Cumulative Probability')
axes[1].set_title(f'Geometric Distribution CDF (p = {p})')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

# Negative Binomial Random Variable
## (Number of experiments until $r$ successes)

```python

# Example: Number of calls until 3rd sale
r_values = [2, 4, 6 , 8]
p_val = 0.3

max_trials = 50

plt.figure(figsize=(10, 3))

for r in r_values:
    x_vals = np.arange(r, max_trials + 1)
    pmf_vals = stats.nbinom.pmf(x_vals - r, r, p_val)
    plt.plot(x_vals, pmf_vals, 'o-', label=f'r = {r}', alpha=0.7)

plt.xlabel('Total Number of Trials')
plt.ylabel('Probability')
plt.title(f'Negative Binomial Distribution for Different r Values (p = {p_val})')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

# Uniform Distribution
## Equally likely to take on any value in its range $(\alpha,\beta
$).

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Parameters for Uniform distribution
a = 2    # Lower bound
b = 7    # Upper bound

print("Uniform Distribution Analysis")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3))

# PDF plot
x_pdf = np.linspace(a-1, b+1, 1000)
y_pdf = stats.uniform.pdf(x_pdf, loc=a, scale=b-a)

ax1.plot(x_pdf, y_pdf,  linewidth=2, label='PDF')
ax1.axhline(y=1/(b-a), color='grey', linestyle='--', alpha=0.7, label=f'Height = {1/(b-a):.2f}')
ax1.fill_between(x_pdf, y_pdf, alpha=0.3)
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.set_title(f'Uniform Distribution PDF\nU({a}, {b})')
ax1.grid(True, alpha=0.3)
ax1.legend()

# CDF plot
x_cdf = np.linspace(a-1, b+1, 1000)
y_cdf = stats.uniform.cdf(x_cdf, loc=a, scale=b-a)

ax2.plot(x_cdf, y_cdf, 'grey', linewidth=2, label='CDF')
ax2.set_xlabel('x')
ax2.set_ylabel('F(x)')
ax2.set_title(f'Uniform Distribution CDF\nU({a}, {b})')
ax2.grid(True, alpha=0.3)
ax2.legend()

plt.tight_layout()
plt.show()
```

# Normal Distribution

```python
mu = 0      # Mean
sigma = 1   # Standard deviation

print("Normal Distribution Analysis")
print(f"Mean (μ) = {mu}")
print(f"Standard deviation (σ) = {sigma}")
print("-" * 50)

# Calculate key probabilities for some z-values
z_values = [-2, -1, 0, 1, 2]

for z in z_values:
    pdf_val = stats.norm.pdf(z, mu, sigma)
    cdf_val = stats.norm.cdf(z, mu, sigma)
    
    print(f'z = {z:2}: PDF = {pdf_val:.4f}, CDF = {cdf_val:.4f}, P(Z > {z}) = {1-cdf_val:.4f}')

print("-" * 50)

# Create the plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3))

# PDF plot
x_pdf = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_pdf = stats.norm.pdf(x_pdf, mu, sigma)

ax1.plot(x_pdf, y_pdf, linewidth=2, label=f'N({mu}, {sigma}²)')
ax1.fill_between(x_pdf, y_pdf, alpha=0.3)
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.set_title(f'Normal Distribution PDF\n(μ = {mu}, σ = {sigma})')
ax1.grid(True, alpha=0.3)
ax1.legend()

# CDF plot
x_cdf = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y_cdf = stats.norm.cdf(x_cdf, mu, sigma)

ax2.plot(x_cdf, y_cdf, 'grey', linewidth=2, label=f'N({mu}, {sigma}²)')
ax2.set_xlabel('x')
ax2.set_ylabel('F(x)')
ax2.set_title(f'Normal Distribution CDF\n(μ = {mu}, σ = {sigma})')
ax2.grid(True, alpha=0.3)
ax2.legend()

plt.tight_layout()
plt.show()
```
